```python
import numpy as np
import os
import cv2
from PIL import Image
import matplotlib.pyplot as plt
initial_path = '../cleandata/(3)/fail'
image_list=os.listdir(initial_path)
n = np.random.randint(len(image_list))
image_path = os.path.join(initial_path, image_list[n])
image = Image.open(image_path)
image_np = np.array(image)
image_np = image_np[240:550, 300:600, :]
print(image_np.shape)
plt.imshow(image_np)
plt.axis('off')
plt.show()
```
